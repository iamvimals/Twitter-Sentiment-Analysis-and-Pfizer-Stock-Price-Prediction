```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sentiment_stock_price =  pd.read_csv("csv/sentiment_score_and_pfizer_price.csv", engine="python")
```

```python
sentiment_stock_price = sentiment_stock_price.rename(columns = {"Unnamed: 0" : "Date"})
```

```python
sentiment_stock_price.set_index(pd.to_datetime(sentiment_stock_price['Date']), inplace = True)
```

```python
sentiment_stock_price = sentiment_stock_price.drop(columns=["Date"])
```

```python
sentiment_stock_price.isnull().sum()
```

```python
sentiment_stock_price.info()
```

```python
sentiment_stock_price.difference_open_close.plot(figsize = (12,8), subplots = True, legend = True);
```

```python
plt.figure(figsize=(16,5), dpi = 100)
plt.plot(sentiment_stock_price.index, sentiment_stock_price.sentiment_1, color='tab:green')
plt.gca().set(title = "Sentiment Scores Time Series", xlabel="Date", ylabel="Sentiment Scores")
plt.show()
```

```python
#sentiment_stock_price.sentiment_2.plot(figsize = (12,8), subplots = True, legend = True);
plt.figure(figsize=(16,5), dpi = 100)
plt.plot(sentiment_stock_price.index, sentiment_stock_price.sentiment_2, color='tab:blue')
plt.gca().set(title = "Sentiment Scores Time Series", xlabel="Date", ylabel="Sentiment Scores")
plt.show()
```

```python
plt.figure(figsize=(16,5), dpi = 100)
plt.plot(sentiment_stock_price.index, sentiment_stock_price["Close"], color='tab:red')
plt.gca().set(title = "Sentiment Scores Time Series", xlabel="Date", ylabel="Sentiment Scores")
plt.show()
```
